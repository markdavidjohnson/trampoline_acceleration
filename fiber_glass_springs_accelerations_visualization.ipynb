```python
path = 'https://raw.githubusercontent.com/markdavidjohnson/trampoline_acceleration/main/fiber_glass_trampoline.csv'
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(path,delimiter=';')
```

```python
df['accel'] = (df['x']**2 + df['z']**2 + df['z']**2)**.5
df['s'] = df['index']/30
```

```python
#df.plot.scatter('index','accel')
#plt.scatter(df['index'], df['accel'])

#plt.show()
scatter=sns.scatterplot(x=df['s'], y=df['accel'],hue=df['accel'])
#scatter.set_xlim(left=2.5, right=20)
scatter.set_xlim(left=11.75, right=13)
plt.rcParams['figure.figsize'] = (15,4)
```
